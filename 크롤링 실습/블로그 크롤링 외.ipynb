#   

##### 1. url 

```python
url = 'https://search.naver.com/search.naver'
params = {
    'where': 'news',
    'query': ''
}
```

##### 2. request  

```python
resp = requests.get(url, params=params)
resp
```

##### 3. resp (soup )

```python
soup = BeautifulSoup(resp.content)
```

```python
ul_tag = soup.find('ul', class_='type01')
```

```python
news_item_tags = ul_tag.find_all('a', class_='_sp_each_title')
len(news_item_tags)
```

```python
news_item_tags[0]
```

#####  a

```python
news_list = []
# {  'title': '',
#     'link': 'url'}
for news in news_item_tags:
    title = news.text
    link = news.get('href')
    news_list.append(
        {
            'title': title,
            'link':link
        }
    )
news_list
```

#####   

```python
li_tags = ul_tag.find_all('li', recursive=False)
len(li_tags)
```

```python
news_list = []
for li_tag in li_tags:
    a_tag = li_tag.find('a', class_='_sp_each_title')
    title = a_tag.text
    link = a_tag.get('href')
    
    dd_tag=li_tag.find_all('dd')[1]
    desc = dd_tag.text
    
    news_list.append({
        'title':title,
        'link':link,
        'desc': desc
    })
news_list
```

#   (5)

```python
# num = 5
page_num = int(input("?"))
query = input("  .")

url = 'https://search.naver.com/search.naver'

result_list = []
for i in range(page_num):
    params = {
        'where': 'post',
        'query': query,
        'start': 1 + i*10
    }
    resp = requests.get(url, params=params)
    
    soup = BeautifulSoup(resp.content)
    ul_tag = soup.find('ul', class_='type01')
    
    blog_list = ul_tag.find_all('li', class_='sh_blog_top')
    
    for post in blog_list:
        dl_tag = post.find('dl')
        title = dl_tag.find('a', class_='_sp_each_title').text
        link = dl_tag.find('a', class_='_sp_each_title').get('href')
        desc = dl_tag.find_all('dd')[1].text
        
        result_list.append({
            'title': title,
            'link': link,
            'desc': desc
        })
    
```
